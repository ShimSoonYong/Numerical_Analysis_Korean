# PROB 2.14

### 문제 2.3 - $ e^{0.5} $의 테일러 급수 근사와 오차 항


#### 1. 테일러 급수 계산

$ x = 0.5 $로 계산하고 $ \sigma(0.5)^3 $까지만 포함하면:
$$
e^{0.5} = 1 + \frac{0.5^1}{1!} + \frac{0.5^2}{2!} + \frac{0.5^3}{3!} + \sigma(0.5)^4
$$
각 항을 계산하면:
$$
e^{0.5} \approx 1 + 0.5 + \frac{0.25}{2} + \frac{0.125}{6} = 1 + 0.5 + 0.125 + 0.0208333 = 1.6458333
$$

#### 2. 오차 항
오차 항은 테일러 급수의 나머지 항
$$
\sigma(0.5)^2 = \frac{(0.5)^4}{4!} \cdot e^{0.5}
$$
여기서 $ e^{0.5} \approx 1.64872 $를 대입:
$$
\sigma(0.5)^2 = \frac{(0.5)^4}{24} \cdot 1.64872 = \frac{0.0625}{24} \cdot 1.64872 \approx 0.004284
$$
따라서 오차 한계는 약 $ 0.004284 $.

### 문제 2.13 - $ \sinh(0.9) $의 테일러 급수 근사와 오차 항

#### 1. 테일러 급수 정의
$ \sinh(x) $의 테일러 급수:
$$
\sinh(x) = \sum_{n=0}^{\infty} \frac{x^{2n+1}}{(2n+1)!}
$$

$ \sigma(x)^5 $까지의 근사를 위해, $ x = 0.9 $일 때:
$$
\sinh(0.9) \approx 0.9 + \frac{(0.9)^3}{3!} + \frac{(0.9)^5}{5!}
$$

$$
\sinh(0.9) \approx 0.9 + \frac{0.729}{6} + \frac{0.59049}{120} = 0.9 + 0.1215 + 0.00492 = 1.02642
$$

#### 2. 오차 항

$ f(x) = \sinh(x) $이고, $ \sinh(x) $의 도함수는 $ \cosh(x) $이다. $ \cosh(0.9) \approx 1.4330864 $이므로:
$$
\sigma(0.9)^5 = \frac{(0.9)^7}{7!} \cdot 1.4330864
$$
$$
\sigma(0.9)^5 \approx \frac{0.4782969}{5040} \cdot 1.4330864 \approx 0.000136
$$
따라서 오차 한계는 약 $ 0.000136 $.

### 문제 2.18 - $ \sin(x) \cos(x) $의 테일러 급수 근사와 오차 항

**문제**: $ \sin(x) \cos(x) $에 대해 $ x = 0 $에서 $ \sigma(x)^5 $까지 근사식을 구하시오.

#### 1. 테일러 급수 정의
$ \sin(x) $와 $ \cos(x) $의 테일러 급수:

$$
\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n+1}}{(2n+1)!}
$$
$$
\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}
$$

이제 두 급수를 곱하여 $ \sin(x) \cos(x) $를 $ \sigma(x)^5 $까지 구한다:
$$
\sin(x) \cos(x) = \left( x - \frac{x^3}{3!} + \frac{x^5}{5!} \right) \times \left( 1 - \frac{x^2}{2!} + \frac{x^4}{4!} \right)
$$
위 곱셈을 전개하면:
$$
\sin(x) \cos(x) \approx x - \frac{x^3}{2} - \frac{x^5}{6}
$$

#### 2. 오차 항
이 근사의 오차 항은 $ \sigma(x)^7 $로 주어진다. 즉, $ n = 2 $에서 발생하는 오차 항은:
$$
\sigma(x)^7 = \frac{|x - a|^{n+1}}{(n+1)!} \cdot \left| \frac{d^{n+1}(\sin(x)\cos(x))}{dx^{n+1}} \right|_{\text{max}}
$$

여기서 $ \sin(x) \cos(x) $의 도함수들을 계산해 최대값을 구할 수 있다.

# PROB 3.24

### 문제 3.24 - 중앙 차분 표현식을 사용

#### 문제:
함수 $ f(x) $가 주어진 점 $ x_i $에서 $ \Delta f_i / h = 0.23751 $이고, $ \nabla f_i / h = 0.24369 $일 때, $ h = 0.1 $입니다. $ f'(x_i) $와 $ f''(x_i) $의 $ \sigma(h^2) $까지의 중앙 차분 표현식에 대한 수치 값을 구하시오.

---


#### 1. 중앙 차분 공식


1. **일차 도함수 $ f'(x_i) $** 에 대한 중앙 차분 공식:
$$
f'(x_i) \approx \frac{f(x_{i+1}) - f(x_{i-1})}{2h}
$$

$$
f'(x_i) \approx \frac{\Delta f_i + \nabla f_i}{2h}
$$
여기서:
- $ \Delta f_i = f(x_{i+1}) - f(x_i) $이고, $ \nabla f_i = f(x_i) - f(x_{i-1}) $

2. **이차 도함수 $ f''(x_i) $** 에 대한 중앙 차분 공식:
$$
f''(x_i) \approx \frac{f(x_{i+1}) - 2f(x_i) + f(x_{i-1})}{h^2}
$$
$$
f''(x_i) \approx \frac{\Delta f_i - \nabla f_i}{h^2}
$$

#### 2. 주어진 값 대입

- $ \Delta f_i / h = 0.23751 $
- $ \nabla f_i / h = 0.24369 $
- $ h = 0.1 $

##### 1차 도함수 $ f'(x_i) $ 계산:
$$
f'(x_i) \approx \frac{(0.23751 \times 0.1) + (0.24369 \times 0.1)}{2 \times 0.1}
$$
$$
f'(x_i) \approx \frac{0.023751 + 0.024369}{0.2} = \frac{0.04812}{0.2} = 0.2406
$$

##### 2차 도함수 $ f''(x_i) $ 계산:
$$
f''(x_i) \approx \frac{(0.23751 \times 0.1) - (0.24369 \times 0.1)}{(0.1)^2}
$$
$$
f''(x_i) \approx \frac{0.023751 - 0.024369}{0.01} = \frac{-0.000618}{0.01} = -0.0618
$$

---

#### 최종 답:
- $ f'(x_i) = 0.2406 $
- $ f''(x_i) = -0.0618 $
